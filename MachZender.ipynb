# P3 
being shifted from $\theta$ = 0 to $\theta$= 360

```python
import numpy as np
from scipy.optimize import curve_fit
from matplotlib import pyplot as plt
from scipy.interpolate import make_interp_spline
V_out_1 = np.array([1.67,1.17,0.65,0.46,0.66,1.23,1.96,2.76,3.22,3.25,3.18,2.90,2.10,1.28,0.73,0.47,0.69,1.15,1.89,2.62,3.03,3.11,2.97,2.63,1.90])
l=len(V_out_1)
#V_out_1=V_out_1.reshape(l,1)
theta_1=np.linspace(0,2*np.pi,num = 25)
theta_1=np.array(theta_1)
%config InlineBackend.figure_format = 'svg'
# For tex fonts and math-mode in figures
plt.rcParams['text.latex.preamble']=r'\usepackage{lmodern}'
params = {'text.usetex':True,'font.size':16,'font.family':
'serif','figure.autolayout': True}
plt.rcParams.update(params)

def test(x, a, b,c,d):
    return a * np.sin(b * x+c)+d
param1, param_cov1 = curve_fit(test, theta_1, V_out_1,p0=[-2,1.5,2,3])
fittedg=param1[0]*np.sin(param1[1]*theta_1+param1[2])+param1[3]

X_Y_Spline = make_interp_spline(theta_1, fittedg)
X_ = np.linspace(theta_1.min(), theta_1.max(), 500)
Y_ = X_Y_Spline(X_)


plt.plot(X_,Y_,color="g")
plt.scatter(theta_1,V_out_1,color="g")
plt.xlabel("$\phi$  /rad")
plt.ylabel("$V_{out}$/ v")
plt.savefig("P3rot.svg")
```
